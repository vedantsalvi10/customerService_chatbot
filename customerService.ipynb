**Set environment variable**

```python
import os

# Set an environment variable
os.environ['GEMINI_API_KEY'] = ''
```

Set the customer service chatbot that response completely on ai bases

```python
from google import genai
from google.genai import types

# The client gets the API key from the environment variable `GEMINI_API_KEY`.
client = genai.Client()
response = client.models.generate_content(
    model="gemini-2.5-flash",
    config=types.GenerateContentConfig(
        system_instruction="You are a polite and friendly AI customer service chatbot.You assist users with general inquiries about a fintech product or service.If you are unsure about an answer, apologize and suggest reaching out to human support.Keep responses short and professional."),
    contents= user_message
)
print(response.text)
```
